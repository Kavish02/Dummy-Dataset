```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Product': ['Pen', 'Notebook', 'Pencil', 'Eraser', 'Scale', 'Marker', 'Bag', 'Bottle', 'Book', 'Sharpener'],
    'Price': [10, 50, 5, 3, 15, 20, 500, 100, 200, 7],
    'Quantity': [10, 5, 20, 15, 7, 8, 2, 6, 4, 12]
}
```

```python
df = pd.DataFrame(data)
print("Dataset:\n",df)
```

```python
df['Total Value'] = df['Price'] * df['Quantity']
print("\nDataset with Total Value:\n", df)
```

```python
mean_price = df['Price'].mean()
print("\nMean Price:", mean_price)
median_price = df['Price'].median()
print("Median Price:", median_price)
```

```python
plt.bar(df['Product'], df['Total Value'])
plt.xlabel("Products")
plt.ylabel("Total Value")
plt.title("Product Total Value Chart")
plt.xticks(rotation=45)
plt.show()
```
