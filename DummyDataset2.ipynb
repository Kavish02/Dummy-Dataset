```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['Amit', 'Riya', 'Karan', 'Sneha', 'Rahul', 'Priya', 'Arjun', 'Neha', 'Vikas', 'Pooja'],
    'Salary': [50000, 60000, None, 55000, 70000, None, 65000, 62000, 72000, 58000]
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Dataset:\n",df)
```

```python
avg_salary = df['Salary'].mean()
print("\nAverage Salary:",avg_salary)
```

```python
print("\nDataset after filling missing values:\n", df)
```

```python
plt.bar(df['Name'], df['Salary'])
plt.xlabel("Employees")
plt.ylabel("Salary")
plt.title("Employee Salary Chart")
plt.xticks(rotation=45)
plt.show()
```
