```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create dataset
data = {
    'Name': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Marks': [80, 60, 90, 70, 85, 50, 78, 65, 92, 15]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df[df['Marks'] > 75])
```

```python
plt.bar(df['Name'], df['Marks'])
plt.show()
```

```python
print(df.isnull())
print(df.isnull().sum())
```
